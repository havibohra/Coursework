```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Load the data from the pickle file
with open('/content/mnist_small.pkl', 'rb') as f:
    data = pickle.load(f)

X = data['X']  # Input features
Y = data['Y']  # Labels

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Perform t-SNE
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
```

```python
# Visualize PCA projection
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
for i in range(10):
    plt.scatter(X_pca[(Y == i).reshape(Y.shape[0]), 0], X_pca[(Y == i).reshape(Y.shape[0]), 1], label=str(i))
plt.title('PCA')
plt.legend()

# Visualize t-SNE projection
plt.subplot(1, 2, 2)
for i in range(10):
    plt.scatter(X_tsne[(Y == i).reshape(Y.shape[0]), 0], X_tsne[(Y == i).reshape(Y.shape[0]), 1], label=str(i))
plt.title('t-SNE')
plt.legend()

plt.tight_layout()
plt.show()
```
